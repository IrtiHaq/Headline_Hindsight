# NMF Topic Modeling

As mentioned, Non-Negative Matrix Factorization (NMF) is an unsupervised machine learning technique that should help us uncover some of main topic among the headlines in our dataset. 

The way our NMF model works is by first treating each headline as a bag of words and then using those it computes the TF-IDF Score for these words. TF-IDF is simply determining how relevant a word is to a given headline by computing the product of many times a word appears in headline and how uncommon a word is (determined by inversing the number of headlines the word appears). The higher the score the more relevant/important a word is for a given headline. The TF-IDF Scores gives us a word vector for each headline, when combined we get matrix with rows of articles on one side and columns corresponding to each word and cells with TDF-Scores. 

Using this TDF-IDF matrix we then simply tell the model to find a certain number of topics, in this case thirty, and then it performs some matrix math wizardry that in the end spits out two other matrixes. One (Word Score Matrix) that assigns each word with a topic score for how closely it is associated with a given topic for each of the topics. The Second matrix (Article Score Matrix) does the same for each individual headline in our dataset. Using the scores on the second matrix we can then sort each article based on their headline into the different topics by assigning it to the topic it has the highest topic score for.

```python
#@title NMF Word Scores
from IPython.display import display

word_table_toggle = widgets.ToggleButton(
    value=False,
    description='Click To Toggle Score',
    disabled=False,
    button_style='info',
    tooltip='Description',
    layout=widgets.Layout(width='200px'),
    icon='check'
)



word_table_label = widgets.HTML(
                value="<h1>NMF Word Score Table</h1><hr>",
            )

word_table_X_header= widgets.HTML(
                value="<p>Color of The Score Corresponds to the level of Association for a Given Topic (i.e. Redder the Color the More Closely Associated the Word is to a Topic))</p>",
            )

box_layout = widgets.Layout(display='flex',
                flex_flow='column',
                align_items='center',
                width='1200px')
cell_hover = {
    "selector": "td:hover",
    "props": [("background-color", "#32CD32")]
}
row_index = {
    'selector':'th.row_heading.level0',
    "props": "background-color: #FF6B6B; color: #192f44; font-weight:normal;"
}
headers = {
    "selector": "th:not(.index_name)",
    "props": "background-color: #192f44; color: white;"
}


def display_word_tbl(data=word_score_all_year_df, show_score=False):
  
  data_new = data.copy().drop('Rank', axis=1)

  if show_score:
    final_table = data_new.shift()[1:].head(10)
    final_table.index.name = 'Rank'
  else:
    word_score_table_1 = data_new.filter(regex='Word', axis=1)
    word_score_table_1.columns = word_score_table_1.columns.str.replace("Word", "")
    word_score_table_1.index.name = 'Rank'

    final_table = word_score_table_1.shift()[1:].head(10)
  
  final_table_styled = final_table.style.format(precision=2) \
                                  .background_gradient(axis=0, cmap='YlOrRd') \
                                  .set_table_styles([cell_hover, 
                                                     headers,
                                                     row_index])

  return display(final_table_styled)

word_table_w = widgets.interactive_output(display_word_tbl,
                                   {'data':widgets.fixed(word_score_all_year_df), 
                                   'show_score':word_table_toggle})

title = widgets.VBox([word_table_label, word_table_toggle, word_table_X_header], layout=box_layout)
table = widgets.VBox([title,word_table_w], layout=widgets.Layout(width='5000px'))
display_table = widgets.VBox([table], layout=widgets.Layout(width='100%', padding='25px'))

display_table
```

The table above was produced using the Word Score Matrix. It shows us the words in a headline that are most associated with a given topic by sorting the words based on their topic scores for a topic (From Highest to Lowest). For each topic, the top ten words with the highest topic scores for a given topic was selected and then sorted from highest to lowest score. Higher Score indicating a higher level of association and thus rank.

By Clicking on the Toggle Score Button, you are then able to see the individual topic scores for each of the words for the specific topic. The color of the score cell is used to encode the magnitude of the score, with a redder color indicating a greater magnitude and thus a higher level of association and a more yellow indicating a weaker score and thus a lower level of association. Color is also used to indicate the current cell being highlighted with green being used and grey highlights words with the same rank as the word that corresponds with the cell being highlighted.

```python
#@title Investigating Topics Over Time
from IPython.display import display

word_score_table = word_score_all_year_df.filter(regex='Word', axis=1)
word_score_table.columns = word_score_table.columns.str.replace("Word", "")

aricle_score_all_year_df['date'] = pd.to_datetime(aricle_score_all_year_df['date'])

def topic_time_graph(topic, year=None,  month=1):
  # Fix Hover Label
  topic = [int(str.split(T)[1]) for T in topic][-5:]
  filt_year = aricle_score_all_year_df[aricle_score_all_year_df['Topic_Asoc'].isin(topic)]

  if year != 'ALL':
    filt_year = filt_year[filt_year['year'] == year]
  
  month = [int(pd.to_datetime(M, format='%b').strftime("%m")) for M in month]
  month = range(month[0], month[1] + 1, 1)

  filt_year = filt_year[(filt_year['month'].isin(month))]

  topic_filt_df = filt_year[['date','Topic_Asoc', 'title']
                            ].groupby([pd.Grouper(key="date", freq="M"),
                                       pd.Grouper(key="Topic_Asoc")]
                                      )['title'].count().reset_index().rename(columns={'title': 'Count'})
  
  topic_filt_df['date'] = topic_filt_df.date - pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)

  graph_1 = px.line(topic_filt_df, 
                  x="date",
                  y="Count", 
                  title='Number of Articles Over Time',
                  color= 'Topic_Asoc',
                  color_discrete_sequence=["#7227ad", "#40688c", "#6ccdd8", 
                                           "#8bcd60", "#f9e42b"],
                  markers=True,
                  width=1000, 
                  height=550,
                  labels={
                     "Count": "Number of Articles",
                     "date": "Month-Year",
                     "Topic_Asoc": "Topic Number"
                 })
  
  graph_1.update_traces(line=dict(width=5),
                      opacity=0.6,
                      marker=dict(size=8,
                                  color='#2a2e36'))
  graph_1.update_xaxes(showgrid=True, gridwidth=3, gridcolor='#E3E6F1', griddash='solid',
                     linewidth=4, linecolor='#6b6b6b')
  graph_1.update_yaxes(showgrid=True, gridwidth=3, 
                     gridcolor='#E3E6F1', griddash='longdash',
                     linewidth=4, linecolor='#6b6b6b')
  graph_1.update_layout(plot_bgcolor='white')
  

  return graph_1.show()



def pub_treemap(topic, month, year=2016):
  
  if year is None:
    return None

  month = [int(pd.to_datetime(M, format='%b').strftime("%m")) for M in month]
  month = range(month[0], month[1] + 1, 1)

  topic = [int(str.split(T)[1]) for T in topic][-5:]
  filt_year = aricle_score_all_year_df

  if year != 'ALL':
    filt_year = filt_year[filt_year['year'] == year]
  
  filt_year_month_topic = filt_year[(filt_year['month'].isin(month)) & 
                                    (filt_year['Topic_Asoc'].isin(topic))]

  data = filt_year_month_topic.groupby(['Topic_Asoc','publication']
                                       )['title'].count().reset_index().rename(columns={'title': 'Count'})

  tmp = px.treemap(data, 
             path=['Topic_Asoc','publication'], 
             values='Count',
             color='publication',
             title='Number of Articles By Publication & Topic',
             custom_data = ['Topic_Asoc','publication', 'Count'])
  tmp.update_traces(marker=dict(cornerradius=5))
  tmp.data[0].hovertemplate = 'Topic %{customdata[0]}<br><em>%{customdata[1]}</em><br>Numer of Articles %{customdata[2]}'
  tmp.data[0].textinfo = 'label+text+value'
  
  tmp.update_layout(
    autosize=False,
    width=600,
    height=600, 
    margin=dict(
        l=5,
        r=20,
        b=5,
        t=70,
        pad=4
    ),
)

  return tmp.show()


def table_color(df):
  table = df.copy()
  colors = ["#f1e4ff", "#e1edfb", "#9ff0ff", "#e1f3d6", "#f9f2b5"]
  topics = list(table['Topic'].unique())

  i = 0
  for topic in topics:
    color = f'background-color: {colors[i]}'
    table.loc[table['Topic'] == topic] = color
    i += 1
  return table


def article_tbl (topic, year=2016, month=1):

  month = [int(pd.to_datetime(M, format='%b').strftime("%m")) for M in month]
  month = range(month[0], month[1] + 1, 1)
  
  topic = [int(str.split(T)[1]) for T in topic][-5:]
  filt_year = aricle_score_all_year_df

  if year != 'ALL':
    filt_year = filt_year[filt_year['year'] == year]
  
  filt_year_month_topic = filt_year[(filt_year['month'].isin(month)) & 
                                    (filt_year['Topic_Asoc'].isin(topic))]

  data = filt_year_month_topic.groupby('Topic_Asoc', group_keys=False) \
                .apply(lambda x: x.sort_values("Max_score", ascending=False)). \
                reset_index(drop=True).groupby('Topic_Asoc', group_keys=False) \
                .head(10).reset_index(drop=True)
  
  data = data[['Topic_Asoc', 'title', 'publication', 
               'Max_score', 'date']
              ].rename(columns={'Topic_Asoc':'Topic', 
                                'title':'Headline',
                                'Max_score':"Topic Score"})
  data["Topic Score"] = data["Topic Score"].round(3)
  data['date'] = data['date'].dt.date

  data_table = data.style \
                   .hide(axis="index") \
                   .apply(table_color, subset=["Topic"], axis=None) \
                   .background_gradient(subset=["Topic Score"], 
                                        cmap='Oranges')
  
  return display(data_table)


topic_selct = widgets.SelectMultiple(
                options=word_score_table.columns,
                value=['Topic 0 '],
                #description='Select Topic (Max 5)',
                rows=7,
                disabled=False,
                layout=widgets.Layout(width='130px'),
                style = {'description_width': 'initial'})
              
year_select = widgets.ToggleButtons(
                options=[None, 2016, 2017, 2018, 2019, 2020, 'ALL'],
                disabled=False,
                button_style='info',
                layout=widgets.Layout(width='200px'),
              )

month_select = widgets.SelectionRangeSlider(
                  options=['Jan', 'Feb', 'Mar' , 'Apr', 'May', 
                           'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 
                           'Nov', 'Dec'],
                  index=(0,11),
                  description='          Months',
                  style = {'description_width': 'initial'},
                  layout=widgets.Layout(width='1100px'),
                  disabled=False
              )

label_Year = widgets.HTML(
                value="<h2>Year</h2><p><em>Select A Year To Start</em></p>",
            )

label_Topic= widgets.HTML(
                value="<h2>Select Topic</h2><p><em>Displays upto 5 Topics</em></p>",
            )
month_label = widgets.HTML(
                value="<h2>Select Months</h2>",
            )

table_title = widgets.HTML(
                value="<h2>Top Headlines By Topic Score</h2>",
            )

w_time_graph = widgets.interactive_output(topic_time_graph, {'topic': topic_selct, 
                                                             'year': year_select,
                                                             'month': month_select})

w_treemap = widgets.interactive_output(pub_treemap, {'topic': topic_selct, 
                                                     'year': year_select, 
                                                     'month': month_select})

w_tbl = widgets.interactive_output(article_tbl, {'topic': topic_selct, 
                                                 'year': year_select, 
                                                 'month': month_select})

side_bar = widgets.VBox([label_Year, year_select, label_Topic, topic_selct])
top_bar = widgets.HBox([side_bar, w_time_graph])

table_bar = widgets.VBox([table_title, w_tbl ], layout= widgets.Layout(height='600px', 
                                                                       overflow_y='auto', 
                                                                       width="650px"))

bottom_bar = widgets.HBox([w_treemap, table_bar], layout=widgets.Layout(width='100%'))

ui = widgets.VBox([top_bar, month_label, month_select, bottom_bar])

display(ui)
```

This Visualization allows us to see how the number of articles by topic varries over time using the line graph (Color of the Line Corresponds to the Topic Number) as well see the  number of publications that cover a certain topic and how many articles they have produced that fall under a certain topic and how that also varries overtime using the treemap. The color of both the Treemap isn't encoding anything in particular but just help to distinguish individual blocks. There is also a table that shows the top 10 headlines for the selected sorted by their topic score. The Color of the Topic Score Cell is used to encode the magnitude to the score, with a higher score coresponding to a darker color. The Color of the Topic Cell Corresponds to Topic and is the same color as the coresponding color of the topic number on the Line Graph.

This means that a publication that publishes a lot of articles overall will dominate most topics simply due to the volume of articles they publish. We will be able to better see basis and get a better sense of the distribution of topics for a publication in the next graph)*




```python
#@title Exploring Publicatins and Topics


data = aricle_score_all_year_df.drop(['date', 'month', 'day', 'title', 'Max_score', 'Topic_Asoc'],axis = 1)
data_num_by_topic = aricle_score_all_year_df[['publication', 
                                              'Topic_Asoc']] \
                                              .groupby(['publication', 
                                                        'Topic_Asoc']) \
                                              .value_counts() \
                                              .unstack(fill_value=0) \
                                              .reset_index()

data_num_by_topic = pd.melt(data_num_by_topic,  id_vars=['publication'],
                            var_name='Topic', value_name='Count')

data_num_by_topic['Percent_Articles'] = (data_num_by_topic['Count']/
                                         data_num_by_topic.groupby('publication')['Count'].transform('sum'))*100

select_publication = alt.selection_point(fields=['publication'])
select_topic = alt.selection_point(fields=['Topic'])

heat_map = alt.Chart(
          data_num_by_topic,
          width=600, height=650,
          title='Exploring the Distribution of Articles By Publications For Each Topic'
          ).mark_bar(
          ).encode(
              alt.Y("publication:N", sort=alt.SortField('Percent_Articles', "descending")),
              alt.X("Topic:O").axis(labelAngle=0, titlePadding = 15),
              alt.Color('Percent_Articles:Q'
              ).scale(scheme='lighttealblue').title('Percent of Articles By The Publication For a Given Topic (%)'),
              tooltip = [alt.Tooltip('Topic'), 
                         alt.Tooltip('publication:N', title='Publication'), 
                         alt.Tooltip('Percent_Articles:Q', 
                                     title='Percent of The Publication Articles (%)',
                                     format=".2f"), 
                         alt.Tooltip('Count:Q', title='Number of Articles', format=".3s")
                         ],
              opacity=alt.condition(select_publication, alt.value(1), alt.value(0.05)),  
              ).add_params(
                  select_publication,
              ).transform_filter(select_topic)

base_pi_chart = alt.Chart(data_num_by_topic, title='Number of Articles (Size) By Topic'
                 ).transform_aggregate(
                    num_by_topic='sum(Count)',
                    groupby=["Topic"]

                 ).encode(
                    alt.Theta("num_by_topic:Q").stack(True),
                    alt.Radius("num_by_topic:Q").scale(type="sqrt", 
                                                zero=True, 
                                                rangeMin=50),
                    color=alt.Color("Topic:O",legend=None
                                    ).scale(
                                        scheme='purplebluegreen'
                                    ),
                    opacity=alt.condition(select_topic, 
                                          alt.value(1), 
                                          alt.value(0.1)),
                    tooltip = ['Topic', 
                               alt.Tooltip('num_by_topic:Q', format=".3s", 
                                           title='Number of Articles')
                              ],
                ).mark_arc(
                    innerRadius=60, stroke="#fff"
                ).add_params(select_topic)

Topic_label = base_pi_chart.mark_text(radiusOffset=30,
                                      angle = 10
                                   ).encode(text="Topic:N",
                                            size = alt.Size(
                                                'num_by_topic:Q',
                                                legend=None
                                                ).scale(
                                                    rangeMin=11
                                                )
                                           )
pi_chat = alt.layer(base_pi_chart, Topic_label)

w = alt.param(name="width", value=400)
h = alt.param(name="height", value=400)

alt.hconcat(
    heat_map, pi_chat.add_params(w, h), center=True
    ).configure_legend(
        orient='bottom',
        gradientLength=600,
        titleFontSize = 13,
        titleLimit=0,
        padding= 10
    )
```

This visualization shows the distribution of Topics among the various news publications from 2016 to 2020 and the Size of the Various Topics. In the Heatmap (Chart to the Left) color is being used to show the percentage of articles a publication publishes that fall under a given topic. A darker Color Indicates that a greater percent of a publication articles falls under a topic. The Second Visualization, the Radial Chart, shows us the size of the different topic. The bars on the Chart Correspond to a certain topic and the size of the bar corresponds to the number of articles that fall under a topic. 

In the visualization the user can select a topic using the Radial Chart this will filter out of the other topics on the Heatmap and then sort the publications by the percent of articles that they have published for the topic Selected. Clicking on a publication on the other hand simply greys out the other publications, allowing the used to be able to focus on a certain publication.
